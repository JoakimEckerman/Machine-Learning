## Function 1:
$g(w) = w \log(w) + (1 - w) \log(1 - w)$

**Solving for Stationary Points:**

We set the derivative $g'(w) = 0$ to find critical points:

$\log(w) - \log(1 - w) = 0$

Solving for $w$:

$w = \frac{1}{2}$

**Evaluating the Second Derivative:**

We evaluate $g''(w)$ at the critical point $w = \frac{1}{2}$:

$g''\left(\frac{1}{2}\right) = 0$

**Nature of the Stationary Point:**

Based on the second derivative test, $g''\left(\frac{1}{2}\right) = 0$, which is inconclusive on its own. However, considering the behavior of the function, at $w = \frac{1}{2}$, the function has a unique value, which is the minimum value within the specified domain $[0, 1]$.

**Checking for Global Minimum:**

To confirm that $w = \frac{1}{2}$ is a global minimum, we can compare the value of the function at this critical point to the values at the endpoints of the domain:

- $g\left(\frac{1}{2}\right)$:

$g\left(\frac{1}{2}\right) = -\log(2) \approx -0.6931$

- $g(0) = 0$
- $g(1) = 0$

Since $g\left(\frac{1}{2}\right) \approx -0.6931$ is the lowest among these values, we can confirm that $w = \frac{1}{2}$ is indeed a global minimum for the function $g(w)$ within the specified domain $[0, 1]$.

## Function 2:
$g(w) = \frac{1}{2} w^T C w + b^T w$ where $C = \begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix}$ and $b = \begin{bmatrix} 1 \\ 1 \end{bmatrix}$

**Gradient ($\nabla g$):**

$\nabla g(w) = \nabla\left(\frac{1}{2}w^T C w + b^T w\right) = \frac{1}{2}(C^T + C)w + b$

**Hessian Matrix ($H$):**

$H(w) = \nabla\nabla g(w) = \frac{1}{2}(C^T + C) = \frac{1}{2}\begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix} = \begin{bmatrix} 1 & \frac{1}{2} \\ \frac{1}{2} & \frac{3}{2} \end{bmatrix}$

**Solving for Stationary Points:**

Set $\nabla g(w) = 0$ to find critical points:

$\frac{1}{2}(C^T + C)w + b = 0$

Substituting the values for $C$ and $b$:

$\frac{1}{2}\begin{bmatrix} 2 & 1 \\ 1 & 3 \end{bmatrix}w + \begin{bmatrix} 1 \\ 1 \end{bmatrix} = 0$

Solving for $w$:

$\begin{bmatrix} 1 & \frac{1}{2} \\ \frac{1}{2} & \frac{3}{2} \end{bmatrix}w + \begin{bmatrix} 1 \\ 1 \end{bmatrix} = 0$

This system of equations can be solved to find the values of $w$. The solution for $w$ is:

$w = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$

**Determine the Nature of the Stationary Point:**

We have found that the stationary point is $w = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$. To determine its nature, we need to calculate the eigenvalues of the Hessian matrix $H(w)$:

- Calculate the eigenvalues of $H(w)$ by solving the characteristic equation $\det(H(w) - \lambda I) = 0$.

The eigenvalues of $H(w)$ are approximately $\lambda_1 \approx 0.5$ and $\lambda_2 \approx 1.5$.

**Nature of the Stationary Point:**

Based on the eigenvalues obtained in step 4:

- Since both eigenvalues $\lambda_1$ and $\lambda_2$ are positive, the stationary point $w = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ is a local minimum.

**Global Minimum (Additional Analysis):**

To check if the local minimum is a global minimum, we can examine the behavior of the function in the surrounding region. Since the Hessian matrix is positive definite (all eigenvalues are positive), the stationary point $w = \begin{bmatrix} 0 \\ 0 \end{bmatrix}$ is indeed a global minimum for the function $g(w)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function 1
def function_1(w):
    return w * np.log(w) + (1 - w) * np.log(1 - w)

# Gradient of Function 1
def gradient_function_1(w):
    return np.log(w) - np.log(1 - w)

# Hessian of Function 1 (second derivative)
def hessian_function_1(w):
    return 1 / w + 1 / (1 - w)

# Solving for Stationary Points of Function 1
def solve_function_1():
    w_critical = 0.5  # Set critical point
    g_double_prime = hessian_function_1(w_critical)
    if g_double_prime > 0:
        return "Local Minimum"
    elif g_double_prime < 0:
        return "Local Maximum"
    else:
        return "Saddle Point (Inconclusive)"

# Function 2
def function_2(w):
    C = np.array([[2, 1], [1, 3]])
    b = np.array([1, 1])
    return 0.5 * w @ C @ w + b @ w

# Gradient of Function 2
def gradient_function_2(w):
    C = np.array([[2, 1], [1, 3]])
    b = np.array([1, 1])
    return C @ w + b

# Hessian of Function 2 (second derivative)
def hessian_function_2(w):
    return np.array([[2, 1], [1, 3]])

# Solving for Stationary Points of Function 2
def solve_function_2():
    w_critical = np.linalg.solve(hessian_function_2(None), -gradient_function_2(np.array([0.0, 0.0])))
    eigenvalues = np.linalg.eigvals(hessian_function_2(w_critical))
    
    if np.all(eigenvalues > 0):
        return "Local Minimum"
    elif np.all(eigenvalues < 0):
        return "Local Maximum"
    else:
        return "Saddle Point (Inconclusive)"

# Evaluate Function 1 and Function 2 at the critical points
critical_point_result_1 = solve_function_1()
critical_point_result_2 = solve_function_2()

print("Function 1:", critical_point_result_1)
print("Function 2:", critical_point_result_2)

# Plot Function 1
w_values = np.linspace(0.01, 0.99, 100)
f1_values = function_1(w_values)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(w_values, f1_values)
plt.title('Function 1: $g(w) = w \log(w) + (1 - w) \log(1 - w)$')
plt.xlabel('w')
plt.ylabel('g(w)')

# Plot Function 2
w1_values = np.linspace(-10, 10, 400)
w2_values = np.linspace(-10, 10, 400)
f2_values = np.zeros((len(w1_values), len(w2_values)))

for i in range(len(w1_values)):
    for j in range(len(w2_values)):
        w_val = np.array([w1_values[i], w2_values[j]])
        f2_values[i, j] = function_2(w_val)

plt.subplot(1, 2, 2)
plt.imshow(f2_values, extent=[-10, 10, -10, 10], origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Function 2: $g(w) = \\frac{1}{2}w^TCw + b^Tw$')
plt.xlabel('w1')
plt.ylabel('w2')

plt.tight_layout()
plt.show()
```

**Derivative of $g(w)$:**

   We have $g(w) = \frac{1}{50}(w^4 + w^2 + 10w)$. To find its derivative $g'(w)$, we'll apply the power rule and constant factor rule for differentiation.

   $$
   \begin{align*}
   g'(w) &= \frac{1}{50} \cdot \frac{d}{dw}(w^4 + w^2 + 10w) \\
   &= \frac{1}{50} \cdot (4w^3 + 2w + 10) \\
   &= \frac{4}{50}w^3 + \frac{2}{50}w + \frac{10}{50} \\
   &= \frac{2}{25}w^3 + \frac{1}{25}w + \frac{1}{5}
   \end{align*}
   $$

So, the derivative of $g(w)$ is $g'(w) = \frac{2}{25}w^3 + \frac{1}{25}w + \frac{1}{5}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(w)
def g(w):
    return (1/50) * (w**4 + w**2 + 10 * w)

# Numerical gradient calculation
def calculate_gradient(g, w, epsilon=1e-6):
    return (g(w + epsilon) - g(w - epsilon)) / (2 * epsilon)

# Gradient Descent Algorithm
def gradient_descent(alpha, max_steps, initial_w):
    weights_history = [initial_w]
    cost_history = [g(initial_w)]
    
    w = initial_w
    
    for _ in range(1, max_steps + 1):
        grad = calculate_gradient(g, w)
        w = w - alpha * grad
        weights_history.append(w)
        cost_history.append(g(w))
    
    return weights_history, cost_history

# Set hyperparameters
alphas = [1, 1e-1, 1e-2]  # Different learning rates
max_steps = 100  # Maximum number of iterations
initial_w = 2.0  # Initial weight

# Run Gradient Descent for different alphas
all_weights_history = []
all_cost_history = []

for alpha in alphas:
    weights_history, cost_history = gradient_descent(alpha, max_steps, initial_w)
    all_weights_history.append(weights_history)
    all_cost_history.append(cost_history)

# Plot the cost function history for each alpha
plt.figure(figsize=(10, 6))

for i, alpha in enumerate(alphas):
    plt.plot(range(len(all_cost_history[i])), all_cost_history[i], label=f'Alpha = {alpha}')

plt.xlabel('Iteration')
plt.ylabel('Cost (g(w))')
plt.title('Cost Function History for Different Learning Rates')
plt.legend()
plt.grid(True)

plt.show()
```

With a step size of 1, the algorithm converges quickly, but it might overshoot the minimum due to its large size.

With a step size of 0.1, the algorithm still converges relatively quickly, and seems less likely to overshoot.

With a step size of 0.01, the algorithm converges more slowly, but it's the least likely to overshoot.

In this particular case, a step size of 0.1 appears to be a reasonable choice as it strikes a balance between convergence speed and stability.
