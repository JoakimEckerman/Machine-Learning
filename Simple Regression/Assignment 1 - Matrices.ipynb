```python
# Author: Joakim Eckerman 028731311
# Date: 08/29/2023
# Course: CECS 456 Machine Learning
# Assignment: Assignment 1

# Description:
# This program calculates Y using given beta values, and the X and noise arrays.
# It then calculates a better estimate for the beta values, prints them and plots the data.

# Instructions:
# 1- Create variables beta_0 and beta_1 such that 𝛽0 = 5 𝑎𝑛𝑑 𝛽1 = 1
# 2- Create an array of observations X of size 100, such that 𝑋𝑖~𝑁(0,1)
# 3- Create an array of noise of size 100, such that 𝜀𝑖~𝑁(0,1)
# 4- Generate the response array Y such that 𝑌 = 𝑏𝑒𝑡𝑎0 +𝑏𝑒𝑡𝑎1 ∗ 𝑋 + 𝑛𝑜𝑖𝑠𝑒
# 5- Create a design matrix X_design (the bold-faced matrix above) such that
# 6- Use the design matrix and the vector Y to find 𝛽

import numpy as np
import matplotlib.pyplot as plt

# Simulated data
beta_0 = 5
beta_1 = 1
X = np.random.normal(0, 1, 100)
noise = np.random.normal(0, 1, 100)
Y = beta_0 + beta_1 * X + noise

# Create design matrix
ones = np.ones(100)
X_design = np.vstack([ones, X]).T

# Calculate beta coefficients
beta = np.dot(np.linalg.inv((np.dot(X_design.T, X_design))), (np.dot(X_design.T, Y)))

# Print calculated beta values
print(beta)

# Plot the data and regression line
plt.scatter(X, Y, label='Data points')
plt.plot(X, beta[0] + beta[1] * X, color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
